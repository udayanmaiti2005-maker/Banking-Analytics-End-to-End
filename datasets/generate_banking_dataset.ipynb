```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta

fake = Faker()
random.seed(42)
np.random.seed(42)

# Configuration
NUM_CUSTOMERS = 5000
NUM_BRANCHES = 10
NUM_ACCOUNTS = 7500
NUM_TRANSACTIONS = 50000
NUM_LOANS = 2000

# Helper functions
cities = ["Kolkata",
    "Delhi",
    "Mumbai",
    "Bangalore",
    "Chennai",
    "Pune",
    "Hyderabad",
    "Ahmedabad"
]

occupations = [ "Engineer",
    "Teacher",
    "Doctor",
    "Business Owner",
    "Student",
    "Government Employee",
    "Private Employee"
]

loan_types = [
    "Home",
    "Car",
    "Education",
    "Personal"
]
```

```python
# Income based balance generation
def generate_balance(income):
  if income < 500000:
    return random.randint(10000, 100000)
  elif income < 1000000:
    return random.randint(50000, 500000)
  else:
    return random.randint(200000, 2000000)

# Credit score generation
def generate_credit_score(income):
  if income < 500000:
    return random.randint(400, 650)
  elif income < 1000000:
    return random.randint(600, 750)
  else:
    return random.randint(700, 850)

# Loan amount generation
def generate_loan(income):
    if income < 500000:
        maximum = 1000000
    elif income < 1000000:
        maximum = 5000000
    else:
        maximum = 10000000
    return random.randint(50000, maximum)

# Transaction amount generation
def generate_transaction_amount(account_type):
    if account_type == "Savings":
      return random.randint(500, 50000)
    else:
      return random.randint(10000, 1000000)

# BRANCH DATA
branches=[]

for i in range(NUM_BRANCHES):
    branches.append({
        "Branch_ID":100+i,
        "Branch_Name":
        f"{cities[i%len(cities)]} Branch",
        "City":
cities[i%len(cities)],
        "Manager_Name":
fake.name()
    })

branches_df=pd.DataFrame(branches)

branches_df.to_csv(
    "Branches.csv",
    index=False
)

# CUSTOMER DATA
customers=[]

for i in range(NUM_CUSTOMERS):
    income=random.randint(
        200000,
        3000000
    )

    customers.append({
        "Customer_ID":
        10001+i,

        "Name":
        fake.name(),
        "Age":
random.randint(
            21,
            70
        ),

        "Gender":
random.choice(
            [
            "Male",
            "Female"
            ]
        ),

        "City":
random.choice(
            cities
        ),

        "Occupation":
random.choice(
            occupations
        ),

        "Annual_Income":
        income,

        "Credit_Score":
generate_credit_score(
            income
        )
    })

customers_df=pd.DataFrame(customers)

customers_df.to_csv(
    "Customers.csv",
    index=False
)

# ACCOUNTS DATA
accounts=[]

for i in range(NUM_ACCOUNTS):

    customer=random.choice(
        customers
    )

    account_type=random.choice(
        [
        "Savings",
        "Current"
        ]
    )

    accounts.append({
        "Account_ID":
        50001+i,

        "Customer_ID":
        customer["Customer_ID"],

        "Branch_ID":
random.randint(
            100,
            109
        ),

        "Account_Type":
        account_type,

        "Balance":
generate_balance(
            customer["Annual_Income"]
        ),

        "Opening_Date":
fake.date_between(
            start_date="-10y",
            end_date="today"
        )
    })

accounts_df=pd.DataFrame(accounts)

accounts_df.to_csv(
    "Accounts.csv",
    index=False
)

# TRANSACTION DATA

transactions=[]

for i in range(NUM_TRANSACTIONS):

    account=random.choice(
        accounts
    )

    transactions.append({
        "Transaction_ID":
        i+1,

        "Account_ID":
        account["Account_ID"],

        "Transaction_Date":
fake.date_between(
            start_date="-3y",
            end_date="today"
        ),

        "Transaction_Type":
random.choice(
            [
            "Deposit",
            "Withdrawal"
            ]
        ),

        "Amount":
generate_transaction_amount(
            account["Account_Type"]
        )
    })

transactions_df=pd.DataFrame(
    transactions
)

transactions_df.to_csv(
    "Transactions.csv",
    index=False
)

# LOAN DATA
loans=[]

for i in range(NUM_LOANS):

    customer=random.choice(
        customers
    )

    loans.append({
        "Loan_ID":
        i+1,

        "Customer_ID":
        customer["Customer_ID"],

        "Loan_Type":
random.choice(
            loan_types
        ),

        "Loan_Amount":
generate_loan(
            customer["Annual_Income"]
        ),

        "Interest_Rate":
round(
            random.uniform(
                5,
                12
            ),
            2
        ),

        "Loan_Status":
random.choice(
            [
            "Approved",
            "Rejected",
            "Pending"
            ]
        )
    })

loans_df=pd.DataFrame(loans)

loans_df.to_csv(
    "Loans.csv",
    index=False
)
```
